````python
import pandas as pd
from sqlalchemy import create_engine
e =create_engine("postgresql+psycopg2://user:123@localhost:5432/Java_imdb_db")
````

```sql
CREATE TABLE name_basics(
nconst char(10) PRIMARY KEY,
primaryname varchar(255),
birthyear char(4),
deathyear CHAR(4),

knownfortitles varchar(255)
);
```

````python
df = pd.read_csv(r".\name.basics.tsv\data.tsv",sep="\t")
df.columns = map(str.lower, df.columns)
````

````python
df.to_sql(
    name="name_basics",
    con=e,
    index=False,
    if_exists='append',
    schema='public')
````
